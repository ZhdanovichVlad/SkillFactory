# Часть 2. Анализ данных

Сразу отметим что, хотя из условий мы должны сделать выборку по всем зимним месяцам 2017 года, по факту в датасете присутствую данные только за январь и февраль.

Т.к. Метрика прибыльности рейсов, предложенная в учебном проекте (стоимость топлива минус стоимость билетов) не отображает все расходы (ЗП, амортизация и т.д.) На основе статей в сети интернет, пришел к выводу что лучше всего использовать следующую метрику. Рейс будет прибыльным если он заполнен на 75%. Себестоимость рейса будет считаться как 75% от стоимости всех билетов на рейсе. Добавить данные признаки.

```python
df['percentage_of_filling'] = df.apply(
    lambda x: x.loc['total_tickets'] / x['total_seats'], axis=1)
```

```python
#  Проверим значение на выбросы
df[(df['percentage_of_filling']>1) | (df['percentage_of_filling']<0)]
```

Выбросов нету.

```python
# Вычислим среднюю стоимость 1 билета
df['cost_for_economy_ticket'] = df.apply(
    lambda x: x['economy_amount'] / x['economy_tickets'], axis=1)
df['cost_for_bis_ticket'] = df.apply(
    lambda x: x['business_amount'] / x['business_tickets'], axis=1)
# Узнаем прибыль самолёта при максимальной загрузке
df['MAX_cost_of_amount'] = df.apply(
    lambda x: x['cost_for_bis_ticket'] * x['business_seats'] +
    x['cost_for_economy_ticket'] * x['economy_seats'], axis=1)
# Вычислим себестоимость каждого полета. 75% от максимальной
df['amount_0,75'] = df.apply(
    lambda x: x['MAX_cost_of_amount'] * 0.75, axis=1)
# Добавим чистую прибыль
df['net_profit'] = df.apply(
    lambda x: x['total_amount'] - x['amount_0,75'], axis=1)
```

Так же предлагаю добавить новый признак. Заполняемость рейсов.
Используем следующую кодировку. 
1 - рейс убыточный (Заполняемость ниже 75%)
2 - рейс себя окупает (Заполняемость выше 75% и ниже 80%)
3 - прибыльный рейс (Заполняемость выше 80% и ниже 90%)
4 - очень прибыльные рейсы (Заполняемость выше 90%)

```python
df['filling_lavel'] = df['percentage_of_filling'].apply(
    lambda x: 'unprofitable' if x < 0.75 else 'low_profitable' if x < 0.8 else 'profitable' if x < 0.9 else 'high_profitable')
```
